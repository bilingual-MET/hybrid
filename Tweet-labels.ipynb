# Tweet-level labels



```python
import pandas as pd
from utilities import Convert, remove_punc
from string import punctuation
```

```python
df =pd.read_csv("data/data_token_labels.csv")
df.head()
```

```python
label = []

for ind, row in df.iterrows():
    list1 = Convert(row['Maori_TokensF'])
    list3 = Convert(row['English_TokensF'])
    list2 = Convert(row['Ambiguous_list'])
    list4 = Convert(row['Unknown_list'])
    X1 = int(len(list1))
    X2 = int(len(list2))
    X3 = int(len(list3))
    X4 = int(len(list4))
  #  print(list1,":",X1," ",list2,":",X2)
    if (X1 == 0 and X2 == 0 and X4 == 0):
        label.append('E') ## English tweet
    elif (X3 == 0 and X2 == 0 and X4 == 0):
        label.append('M') 
    elif (X2 == 0 and X4 == 0):
        label.append('B') ## Bilingual 
    else:
        label.append("U")


df['Tweet_labels'] = label  
df
```

```python
### additional check. 


import string
import os
from tensorflow import keras
import numpy as np
import tensorflow as tf
import re
import pickle
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from utilities import Convert, remove_punc
config = tf.compat.v1.ConfigProto()
config.gpu_options.allow_growth = True
sess = tf.compat.v1.Session(config=config)
```

```python
## classes are [0] for Bilingual; [1] for Maori; [2] for English (these are classes as per James et al., 2022)

model_path = "BiGRU-attention-miwiki_model2_MPW300SG.h5"
tokenizer_path = "tokenizerBiGRU-attention-miwiki_model2_MPW300SG.pickle"

```

```python
loaded_model = tf.keras.models.load_model(model_path)
```

```python
loaded_model.summary()
```

```python
## loading tokenizer. 
with open(tokenizer_path, 'rb') as handle:
        tokenizer = pickle.load(handle)
```

```python
### best to run using a GPU machine

label = []
for ind, row in df.iterrows():
    x = row['modified_text']
    seq= tokenizer.texts_to_sequences([x]) 
    padded = pad_sequences(seq, maxlen=250)
    predict=loaded_model.predict(padded)
    classw=np.argmax(predict,axis=1)
    cll = int(classw[0]) ### Predicting class label based on the model. 
    if cll == 2:
        l = 'E'
        label.append(l) ## English
    elif cll == 1:
        l = 'M'
        label.append(l) ## Maori
    elif cll == 0:
        l = 'B'
        label.append(l) ## Maori
    else:
        l= 'U'
        label.append(l)

df['ML_Tweet_labels'] = label        
```

```python
checked = []

for ind, row in df.iterrows():
    if row['Tweet_labels'] == row['ML_Tweet_labels']:
        checked.append("yes")
    else:
        checked.append("no")

df['Correct_Tweet_label'] = checked

df.head()
```
